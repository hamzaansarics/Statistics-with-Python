### Q Test or Q Dixon Test
- they are used to check whehter outliers are present or not in our data
- first we will state our hypothesis as
- H0 ( There is no outliers in our data
- H1 ( There is outlier exist in our data)
- formula is: d = gap/range
- if d>(table_value) Reject the H0
- if d<(table_value) Accept the H0

```python
df=pd.read_csv(r'I:\Text Datasets\CSV\Iris.csv').sample(10)
```

```python
confidence_level=0.95
N = df.shape[0]
```

```python
#1 Stating Hypothesis
H0="There is no Outliers in Data"
H1="There is Outlier in Data"
```

```python
#2 Sorting Data 
sorted_df=df['SepalLengthCm'].sort_values()
```

```python
gap=sorted_df.iloc[0]-sorted_df.iloc[1]
```

```python
range_=sorted_df.max()-sorted_df.min()
```

```python
dixon=gap/range_
```

```python
# Our critical Value is on 0.95 Confidence level
critical_value=0.466
```

```python
if dixon > critical_value:
    print('Reject the H0 Hypothesis')
if dixon<= critical_value:
    print('Accept the H0 Hypothesis')
```
