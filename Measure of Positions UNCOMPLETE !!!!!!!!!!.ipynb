#### Standard Score
- They are used to find the position of data element that how many standard deviation those point are away from mean
- (x - MEAN)/std

```python
# We are chosing any data point from duration feature of ted talks datasets
(1190-df['duration'].mean())/df['duration'].std()
```

```python
#the result which shows 0.98 standard deviation data point away from mean in positive and the data point
# exist in 68%  around the mean as shown in image 34.13% | 34.13%
```

#### Percentile

```python
np.percentile(df.duration,[0,1,2,3,4,5])
```
