### QQ Plot
- they are used to check that our sample data come from a population which is following normal distribution or not 
- If the line between Thoretical and Quantile plots its mean our data come from Noramlly distribution

### Example 1 for Normally Distributed

```python
sns.distplot(df['SepalWidthCm'])
```

```python
sample=df[['SepalWidthCm']].sample(50).sort_values('SepalWidthCm')
```

```python
sample['i']=np.linspace(1,sample.shape[0],sample.shape[0])
```

```python
sample
```

```python
sample['(i-0.5)/n']=(sample['i']-0.5)/sample.shape[0]
```

```python
sample
```

```python
sample['ZScore']=(df['SepalWidthCm']-df['SepalWidthCm'].mean())/df['SepalWidthCm'].std()
```

```python
# Importing to Excel to find ZT Score
sample.to_csv('I:\Text Datasets\CSV\QQ.csv')
```

### Example 2 for Non-Normally Distributed

```python
sns.distplot(df['PetalLengthCm'])
```

```python
sample_n=df[['PetalLengthCm']].sample(50).sort_values('PetalLengthCm')
```

```python
sample_n['i']=np.linspace(1,50,50)
```

```python
sample_n['(i-0.5)/n']=(sample_n['i']-0.5)/sample_n.shape[0]
```

```python
sample_n['ZScore']=(sample_n['PetalLengthCm']-sample_n['PetalLengthCm'].mean())/sample_n['PetalLengthCm'].std()
```

```python
sample_n.to_csv('I:\Text Datasets\CSV\QQ_n.csv',index=False)
```
