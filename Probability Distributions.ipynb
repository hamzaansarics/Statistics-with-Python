#### Binoial Distribution
- a type of distribution that gets from binomial experiment
    - Binomial Experiment
        - Number of Trial should be fixed
        - Number of outcomes should be 2
        - Probability of events should be same
        - Events should be independent
    - P(Bin) = n!/(n-x)!x!*(p)^x(q)^n-x
        - n the number of trials
        - x the number of success from whole tirals
        - p the probability of getting success
        - q is the failure chances

#### Geometric Distribution
- they are used to give the probability of success after failures
- f(x) = (1  p)x  1p
- The three assumptions are:

    - There are two possible outcomes for each trial (success or failure).
    - The trials are independent.
    - The probability of success is the same for each trial.
- Sample question: If your probability of success is 0.5, what is the probability you meet an independent voter on your third try?

##### As the x values increases probability going down

#### Hypergeometric Distribution
- when events are not independent then we used this 
- they are similar to binomial distribution when sample is less than 5% because they give result similar to bino. and also they used combination formula


```python
Image(r'C:\Users\H A M Z A\Pictures\hyper.png')
```

```python
import matplotlib.pyplot as plt
[M, n, N] = [20, 7, 12]
rv = stat.hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_dogs = rv.pmf(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_dogs, 'bo')
ax.vlines(x, 0, pmf_dogs, lw=2)
ax.set_xlabel('# of dogs in our group of chosen animals')
ax.set_ylabel('hypergeom PMF')
plt.show()
```
