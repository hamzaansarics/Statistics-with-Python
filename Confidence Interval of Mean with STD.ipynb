### Confidence Interval
- they are used to find the range interval for population mean 
- when we find the sample mean its not possible to know how much our sample mean is close to populaton mean so statistican drive a interval whcih shows P_Mean will be exist between these interval
- Formula is below this cell where 90% confidence shows its own value and etc

##### They can be find when standard deviation of population is given

```python
df=pd.read_csv('I:\Text Datasets\CSV\yahoo_finance.csv')
```

```python
df.head()
```

```python
# we are taking only 500 sample of data for analysis to check whether this will work or not
sample=df.Open.sample(500)
```

```python
# real data size 
df.shape[0]
```

```python
conf_interval=95.0
std=df['Open'].std()
X=sample.mean()
```

```python
# when confidence value is 95.0 then value will be 1.96 as shown in image
#Za/2=1.96 
```

```python
min_interval=X-1.96*(std/np.sqrt(sample.shape[0]))
```

```python
max_interval=X+1.96*(std/np.sqrt(sample.shape[0]))
```
